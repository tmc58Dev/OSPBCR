```python
import os

project_name = "OS-PBCR"

folders = [
    # Assets
    "assets/css/global",
    "assets/css/components",
    "assets/css/pages",

    "assets/js/core",
    "assets/js/pages",

    "assets/images/logos",
    "assets/images/hero",
    "assets/images/team",
    "assets/images/participants",
    "assets/images/misc",

    "assets/pdf/angul",
    "assets/pdf/balangir",
    "assets/pdf/bargarh",
    "assets/pdf/bhadrak",

    "assets/svg",

    "assets/data",

    "assets/libraries/bootstrap",
    "assets/libraries/gsap",
    "assets/libraries/fontawesome",

    "assets/videos",

    # Components
    "components",

    # Docs
    "docs"
]

files = [
    # Root HTML Pages
    "index.html",
    "about.html",
    "participants.html",
    "training-materials.html",
    "data-entry-status.html",
    "data-sources.html",
    "team.html",
    "map.html",

    # CSS
    "assets/css/main.css",

    # JS
    "assets/js/main.js",

    # SVG
    "assets/svg/odisha-map.svg",

    # JSON Data
    "assets/data/districts.json",
    "assets/data/participants.json",
    "assets/data/team.json",
    "assets/data/hospitals.json",
    "assets/data/training-materials.json",
    "assets/data/population-data.json",
    "assets/data/cancer-cases.json",
    "assets/data/dashboard.json",

    # Components
    "components/navbar.html",
    "components/footer.html",
    "components/hero.html",
    "components/district-card.html",
    "components/participant-card.html",
    "components/team-card.html",
    "components/data-table.html",
    "components/map-tooltip.html",

    # Root Files
    "README.md",
    "favicon.ico"
]

# Create root project folder
os.makedirs(project_name, exist_ok=True)

# Create folders
for folder in folders:
    os.makedirs(os.path.join(project_name, folder), exist_ok=True)

# Create files
for file in files:
    path = os.path.join(project_name, file)

    if not os.path.exists(path):
        with open(path, "w", encoding="utf-8") as f:
            if file.endswith(".html"):
                f.write(f"<!-- {os.path.basename(file)} -->\n")
            elif file.endswith(".css"):
                f.write("/* Main Styles */\n")
            elif file.endswith(".js"):
                f.write("// Main JavaScript\n")
            elif file.endswith(".json"):
                f.write("{}")
            elif file.endswith(".md"):
                f.write("# OS-PBCR\n")

print(f"✅ '{project_name}' project structure created successfully!")
```

```python
import os

def generate_tree(root_dir=".", prefix=""):
    entries = sorted(
        [e for e in os.listdir(root_dir) if not e.startswith(".")]
    )

    for i, entry in enumerate(entries):
        path = os.path.join(root_dir, entry)
        is_last = i == len(entries) - 1

        connector = "└── " if is_last else "├── "
        print(prefix + connector + entry)

        if os.path.isdir(path):
            extension = "    " if is_last else "│   "
            generate_tree(path, prefix + extension)

if __name__ == "__main__":
    print(os.path.abspath("."))
    generate_tree(".")
```
